# 04 Operators

Hands-on examples of arithmetic, comparison, logical, assignment, identity, membership and bitwise operators â€” key use-case: perform calculations, compare values for conditional logic, and implement simple business rules.

## Arithmetic Operators

```python
# Basic arithmetic operations
a = 10
b = 3

print("=== Arithmetic Operators ===")
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
```

## Comparison Operators

```python
# Comparing values
a = 10
b = 3

print("=== Comparison Operators ===")
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")
print(f"{a} > {b}: {a > b}")
print(f"{a} < {b}: {a < b}")
print(f"{a} >= {b}: {a >= b}")
print(f"{a} <= {b}: {a <= b}")
```

## Logical Operators

```python
# Boolean logic operations
x = True
y = False

print("=== Logical Operators ===")
print(f"x and y: {x and y}")
print(f"x or y: {x or y}")
print(f"not x: {not x}")
```

## Assignment Operators

```python
# Shorthand assignment operations
num = 5
print(f"Initial value: {num}")

num += 3  # num = num + 3
print(f"After += 3: {num}")

num -= 2  # num = num - 2
print(f"After -= 2: {num}")

num *= 2  # num = num * 2
print(f"After *= 2: {num}")

num /= 3  # num = num / 3
print(f"After /= 3: {num}")
```

## Identity Operators

```python
# Check if objects are the same (not just equal)
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print("=== Identity Operators ===")
print(f"list1 is list3: {list1 is list3}")
print(f"list1 is list2: {list1 is list2}")
print(f"list1 == list2: {list1 == list2}")
print(f"list1 is not list2: {list1 is not list2}")
```

## Membership Operators

```python
# Check if value exists in a sequence
fruits = ["apple", "banana", "orange"]

print("=== Membership Operators ===")
print(f"'apple' in fruits: {'apple' in fruits}")
print(f"'grape' in fruits: {'grape' in fruits}")
print(f"'grape' not in fruits: {'grape' not in fruits}")
print(f"'banana' not in fruits: {'banana' not in fruits}")
```

## Bitwise Operators

```python
# Operations on binary representations
a = 10  # Binary: 1010
b = 4   # Binary: 0100

print("=== Bitwise Operators ===")
print(f"a & b (AND): {a & b}")      # 0000 = 0
print(f"a | b (OR): {a | b}")       # 1110 = 14
print(f"a ^ b (XOR): {a ^ b}")      # 1110 = 14
print(f"~a (NOT): {~a}")            # Inverts bits
print(f"a << 2 (Left Shift): {a << 2}")   # 101000 = 40
print(f"a >> 2 (Right Shift): {a >> 2}")  # 10 = 2
```
