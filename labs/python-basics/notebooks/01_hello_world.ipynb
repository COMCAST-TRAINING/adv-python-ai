## Basic Print Statement

```python
# The classic first program in any languageprint("Hello, World!")
```

## Print Multiple Lines

```python
# Print multiple stringsprint("Hello, Python!")print("Welcome to programming!")
```

## Print with Separator

```python
# Use sep parameter to change separatorprint("Hello", "Python", "World", sep=", ")
```

## Print with Custom End

```python
# Use end parameter to change line endingprint("Hello", end=" ")print("World!")
```

## String Concatenation

```python
# Combine strings with +greeting = "Hello"name = "Python"print(greeting + ", " + name + "!")
```

## F-String Formatting

```python
# Modern Python string formatting with f-stringsname = "Alice"age = 25print(f"My name is {name} and I am {age} years old")
```
