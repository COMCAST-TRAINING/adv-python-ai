# 01 Hello World

Get started with Python: write and run your first script, use `print()` to display output, and learn basic syntax and indentation. Key use case: verify the Python environment and test simple scripts.

## Basic Print Statement

```python
# The classic first program in any language
print("Hello, World!")
```

## Print Multiple Lines

```python
# Print multiple strings
print("Hello, Python!")
print("Welcome to programming!")
```

## Print with Separator

```python
# Use sep parameter to change separator
print("Hello", "Python", "World", sep=", ")
```

## Print with Custom End

```python
# Use end parameter to change line ending
print("Hello", end=" ")
print("World!")
```

## String Concatenation

```python
# Combine strings with +
greeting = "Hello"
name = "Python"
print(greeting + ", " + name + "!")
```

## F-String Formatting

```python
# Modern Python string formatting with f-strings
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old")
```
