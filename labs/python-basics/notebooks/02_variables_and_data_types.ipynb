# 02 Variables And Data Types

Create and manipulate variables and Python's core data types (int, float, str, bool, None). Key use case: store program state or configuration values, parse user input, and prepare data for computation and function calls.

## Variable Assignment

```python
# Variables are containers for storing data
name = "John"
age = 30
height = 5.9
is_student = False
print(f"Name: {name}, Age: {age}, Height: {height}, Student: {is_student}")
```

## Multiple Assignment

```python
# Assign multiple variables in one line
x, y, z = 1, 2, 3
print(f"x={x}, y={y}, z={z}")

# Assign same value to multiple variables
a = b = c = 10
print(f"a={a}, b={b}, c={c}")
```

## Integer Type

```python
# Whole numbers without decimal points
count = 42
negative = -15
large_num = 1_000_000  # Underscores for readability
print(f"Count: {count}, Type: {type(count)}")
print(f"Large number: {large_num}")
```

## Float Type

```python
# Numbers with decimal points
price = 19.99
pi = 3.14159
scientific = 1.5e3  # 1.5 × 10³ = 1500
print(f"Price: {price}, Type: {type(price)}")
print(f"Scientific: {scientific}")
```

## String Type

```python
# Text data enclosed in quotes
single = 'Hello'
double = "World"
multiline = """This is
a multiline
string"""
print(f"Single quotes: {single}")
print(f"Multiline:")
print(multiline)
```

## Boolean Type

```python
# True or False values
is_active = True
is_deleted = False
print(f"Active: {is_active}, Type: {type(is_active)}")
print(f"5 > 3 = {5 > 3}")  # Comparison returns boolean
```

## None Type

```python
# Represents absence of value
result = None
print(f"Result: {result}, Type: {type(result)}")

# Commonly used for optional parameters
def greet(name=None):
    if name is None:
        return "Hello, Guest!"
    return f"Hello, {name}!"

print(greet())
print(greet("Alice"))
```

## Type Conversion

```python
# Convert between data types
str_num = "100"
num = int(str_num)  # String to integer
float_num = float(str_num)  # String to float
print(f"String: {str_num}, Integer: {num}, Float: {float_num}")

# Convert to string
age = 25
age_str = str(age)
print(f"Age as string: '{age_str}', Type: {type(age_str)}")
```

## Type Checking

```python
# Check the type of variables
value = 42
print(f"Type of {value}: {type(value)}")
print(f"Is integer? {isinstance(value, int)}")
print(f"Is string? {isinstance(value, str)}")
```
