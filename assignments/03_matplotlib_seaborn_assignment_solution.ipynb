````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load the 'penguins' dataset
penguins = sns.load_dataset('penguins')

# Display the first few rows
print("First 5 rows of the Penguins dataset:")
print(penguins.head())

# Display dataset info
print("\nDataset Info:")
print(penguins.info())
````

## Exercise 1: Scatter Plot (Relational)
Create a scatter plot showing the relationship between `bill_length_mm` (x-axis) and `flipper_length_mm` (y-axis).
- Use Seaborn's `scatterplot`.
- Color the points based on the `species` column (hue).

````python
# Create a scatter plot of flipper_length_mm vs body_mass_g

plt.figure(figsize=(10, 6))
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species', style='sex', s=100)
plt.title('Flipper Length vs Body Mass of Penguins')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.legend(title='  Species / Sex', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
# Show the plot
plt.tight_layout()
plt.show()  
````

### Interpretation:
- **Positive Correlation**: There is a clear positive correlation between flipper length and body mass; as flipper length increases, body mass tends to increase.
- **Species Separation**: 
    - **Gentoo** penguins are distinctively larger, occupying the top-right portion of the plot (high flipper length, high body mass).
    - **Adelie** and **Chinstrap** penguins are smaller and cluster in the bottom-left.
- **Sexual Dimorphism**: Within each species cluster, there is a separation by sex (indicated by marker style), with males generally being larger and heavier than females.

## Exercise 2: Histogram (Distribution)
Visualize the distribution of the `body_mass_g` column.
- Use Seaborn's `histplot`.
- Add a Kernel Density Estimate (KDE) line.
- Set the number of bins to 20.

````python
# Create a histogram of body_mass_g
plt.figure(figsize=(10, 6))
sns.histplot(data=penguins, x='body_mass_g', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Body Mass of Penguins')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
# Show the plot
plt.tight_layout()
plt.show()
````

### Interpretation:
- **Bimodal Distribution**: The histogram (and the KDE line) suggests a bimodal distribution because **there are two distinct "humps" or peaks separated by a valley**. 
    - If the data were unimodal (like a normal distribution), you would see a single bell curve. 
    - Here, the dip between the two peaks indicates that the data likely comes from two different underlying groups mixed together.
- **Range**: The body mass of the penguins ranges from approximately 2,700g to 6,300g.
- **Peaks**:
    - The first, larger peak is centered around 3,500g - 4,000g, representing the smaller species.
    - The second, smaller peak is around 5,000g - 5,500g, representing the larger species.

## Exercise 3: Bar Plot (Categorical)
Create a bar plot showing the average `body_mass_g` for each `species`.
- Use Seaborn's `barplot`.
- Show error bars (default behavior).

````python
# Create a bar plot of average body_mass_g by species
plt.figure(figsize=(8, 6))
sns.barplot(data=penguins, x='species', y='body_mass_g', palette='pastel', hue='species')
plt.title('Average Body Mass of Penguins by Species')
plt.xlabel('Species')
plt.ylabel('Average Body Mass (g)')
plt.grid(True, axis='y', alpha=0.3)
# Show the plot
plt.tight_layout()
plt.show() 
````

### Interpretation:
- **Heaviest Species**: **Gentoo** penguins are clearly the heaviest on average, with a mean body mass exceeding 5,000g.
- **Similar Sizes**: **Adelie** and **Chinstrap** penguins have very similar average body masses, both hovering around 3,700g.
- **Error Bars**: The vertical black lines at the top of each bar represent the **confidence interval** (typically 95%). 
    - They show the uncertainty in the estimate of the mean.
    - Since the error bars for Adelie and Chinstrap overlap slightly (or are very close), their average masses might not be statistically significantly different without further testing, whereas Gentoo is distinctly different.

## Exercise 4: Box Plot (Categorical Distribution)
Create a box plot to compare the distribution of `bill_depth_mm` across different `island`s.
- Use Seaborn's `boxplot`.
- Differentiate by `sex` using `hue`.

````python
# box plot to compare the distribution of `bill_depth_mm` across different `island`s
plt.figure(figsize=(8, 6))
# Updated hue to 'sex' to differentiate by gender as per instructions
sns.boxplot(data=penguins, x='island', y='bill_depth_mm', palette='Set2', hue='sex')
plt.title('Bill Depth Distribution across Islands by Sex')
plt.xlabel('Island')
plt.ylabel('Bill Depth (mm)')
plt.grid(True, axis='y', alpha=0.3)
# Show the plot
plt.tight_layout()  
plt.show()
````

### Interpretation:
- **Island Differences**: 
    - **Biscoe** island has a noticeably lower median bill depth compared to **Dream** and **Torgersen**. This is influenced by the species distribution (Biscoe hosts Gentoos which have shallower bills).
    - **Dream** and **Torgersen** have similar, higher bill depth distributions.
- **Sexual Dimorphism**: 
    - Within each island, there is a clear separation between sexes. **Males** (typically the higher box) have deeper bills than **females**.
- **Box Components**:
    - The **box** represents the Interquartile Range (IQR), containing the middle 50% of the data.
    - The **horizontal line** inside the box is the median.
    - The **whiskers** extend to the min/max values within 1.5 * IQR.
- **Outliers**:
    - Any individual points (diamonds) that appear **above or below the whiskers** are considered outliers. These are data points that fall statistically far from the rest of the distribution.

## Exercise 5: Heatmap (Correlation)
Visualize the correlation between numerical variables in the dataset.
1. Calculate the correlation matrix using `penguins.corr(numeric_only=True)`.
2. Create a heatmap using Seaborn's `heatmap`.
3. Annotate the cells with the correlation values.
4. Use a diverging color map (e.g., 'coolwarm').

````python
# Calculate the correlation matrix
corr = penguins.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Draw the heatmap
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Penguin Variables')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()  # Adjusts the plot to ensure everything fits without overlapping

# Show the plot
plt.show()
````

### Interpretation:
- **Strong Positive Correlation**: 
    - `flipper_length_mm` and `body_mass_g` have a very strong positive correlation (around **0.87**). This makes sense biologically: larger penguins (heavier) tend to have longer flippers to propel themselves.
    - `bill_length_mm` also correlates positively with `flipper_length_mm` and `body_mass_g`.
- **Negative Correlation**:
    - `bill_depth_mm` has a negative correlation with `flipper_length_mm` (**-0.58**) and `body_mass_g` (**-0.47**). This is interesting because Gentoo penguins (the largest species) actually have shallower bills compared to the smaller Adelie and Chinstrap penguins.
- **Diagonal**: The diagonal values are always **1.00** because every variable is perfectly correlated with itself.

### Code Explanation (Ticks):
- **`plt.xticks(rotation=45, ha='right')`**: This rotates the labels on the x-axis by 45 degrees and aligns them to the right. This is crucial when labels are long (like variable names) to prevent them from overlapping with each other.
- **`plt.yticks(rotation=0)`**: This ensures the labels on the y-axis remain horizontal (0 degrees). This is standard for heatmaps as it keeps the row names easily readable from left to right.

## Exercise 6: Pair Plot (Multivariate)
Create a pair plot to visualize pairwise relationships in the entire dataset.
- Use Seaborn's `pairplot`.
- Color the points by `species`.

````python
# Create a pair plot of the entire dataset
g = sns.pairplot(penguins, hue='species', palette='husl', markers=["o", "s", "D"])
g.figure.suptitle('Pairwise Relationships in the Penguin Dataset', fontsize=16, y=1.02)

# Show the plot
plt.show()
````

### Interpretation:
- **Overview**: A pair plot displays pairwise relationships between all numerical variables in the dataset. The diagonal shows the distribution of a single variable, while the off-diagonal plots show scatter plots between two variables.
- **Diagonal (Distributions)**:
    - The diagonal plots (KDEs) show how each variable is distributed for each species.
    - You can see that **Gentoo** (green) is often separated from the other two species, especially in `flipper_length_mm` and `body_mass_g`.
- **Scatter Plots (Relationships)**:
    - **Clusters**: The scatter plots reveal distinct clusters. For example, plotting `bill_length_mm` vs. `bill_depth_mm` shows three fairly distinct groups corresponding to the three species.
    - **Correlations**: You can visually confirm the correlations seen in the heatmap. For instance, the strong positive linear relationship between `flipper_length_mm` and `body_mass_g` is clearly visible as an upward sloping line of points.

## Exercise 7: Customizing Plots with Matplotlib
Create a figure with two subplots (1 row, 2 columns) using `plt.subplots()`.
1. In the first subplot, create a count plot of `island`.
2. In the second subplot, create a scatter plot of `bill_length_mm` vs `bill_depth_mm`.
3. Set a title for the entire figure ("Penguin Analysis").
4. Set titles for each subplot.

````python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# First subplot: Count plot of island
sns.countplot(data=penguins, x='island', palette='pastel', ax=axs[0], hue='island')
axs[0].set_title('Count of Penguins by Island')
axs[0].set_xlabel('Island')
axs[0].set_ylabel('Count')
axs[0].grid(True, alpha=0.3)

# Second subplot: Scatter plot of bill_length_mm vs bill_depth_mm
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', style='sex', s=100, ax=axs[1])
axs[1].set_title('Bill Length vs Bill Depth of Penguins')
axs[1].set_xlabel('Bill Length (mm)')
axs[1].set_ylabel('Bill Depth (mm)')
axs[1].grid(True, alpha=0.3)

# Overall title for the figure
plt.suptitle('Penguin Analysis', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to make room for the suptitle

# Show the plot
plt.show()
````

### Interpretation:
- **Subplot 1 (Left): Count of Penguins by Island**
    - This bar chart shows the distribution of data collection across locations.
    - **Biscoe** island has the highest number of observations in this dataset.
    - **Torgersen** has the fewest observations.
- **Subplot 2 (Right): Bill Length vs. Bill Depth**
    - This scatter plot reveals distinct clusters for the three species based on beak shape.
    - **Adelie** (blue): Short bill length, high bill depth.
    - **Gentoo** (green): Long bill length, low bill depth.
    - **Chinstrap** (orange): Long bill length, high bill depth.
    - This plot effectively demonstrates how using two physical measurements can help distinguish between different species better than using just one.
